```python
import json
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor

# Function to parse a single JSON line
def parse_json_line(line):
    try:
        return json.loads(line)
    except json.JSONDecodeError as e:
        print(f"Error decoding line: {e}")
        return None

# Function to load JSON objects in parallel
def load_json_to_dict_parallel(file_path, num_workers=4):
    data = []
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with ThreadPoolExecutor(max_workers=num_workers) as executor:
        results = list(tqdm(executor.map(parse_json_line, lines), total=len(lines), desc="Processing Lines", unit="line"))

    # Filter out None values from failed parses
    data = [result for result in results if result is not None]
    return data

# Example usage
file_path = 'ecg_signals.json'
json_dict_list = load_json_to_dict_parallel(file_path)

if json_dict_list:
    print(json_dict_list[0])
```

```python
import neurokit2 as nk
import numpy as np
from scipy import signal
from scipy.stats import skew, kurtosis

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    """
    Apply a bandpass filter to the input signal

    Parameters:
    - data: Input signal
    - lowcut: Lower cutoff frequency
    - highcut: Higher cutoff frequency
    - fs: Sampling frequency
    - order: Filter order

    Returns:
    - Filtered signal
    """
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.butter(order, [low, high], btype='band')
    y = signal.filtfilt(b, a, data)
    return y

def extract_comprehensive_ecg_features(filtered_signal, fs=250):
    """
    Robust ECG feature extraction with comprehensive analysis

    Parameters:
    - filtered_signal: Preprocessed ECG signal
    - fs: Sampling frequency

    Returns:
    - Dictionary of extracted ECG features
    """
    try:
        # Clean the signal
        ecg_cleaned = nk.ecg_clean(filtered_signal, sampling_rate=fs)

        # Detect R peaks with multiple methods
        try:
            signals, info = nk.ecg_peaks(ecg_cleaned, sampling_rate=fs)
        except Exception:
            # Fallback to alternative peak detection
            try:
                signals, info = nk.ecg_peaks(ecg_cleaned, sampling_rate=fs, method='hamilton')
            except Exception:
                raise ValueError("Unable to detect R-peaks")

        # Ensure R peaks are detected
        r_peaks = info.get("ECG_R_Peaks", [])
        if len(r_peaks) < 2:
            raise ValueError(f"Insufficient R-peaks detected. Found {len(r_peaks)} peaks")

        # Features dictionary
        features = {}

        # 1. Temporal Domain Features
        if len(r_peaks) > 1:
            # R-R Interval Analysis
            rr_intervals = np.diff(r_peaks) / fs  # Convert to seconds
            features.update({
                "rr_interval_mean": np.mean(rr_intervals),
                "rr_interval_std": np.std(rr_intervals),
                "rr_interval_min": np.min(rr_intervals),
                "rr_interval_max": np.max(rr_intervals),
            })

            # Heart Rate Metrics
            heart_rates = 60 / rr_intervals
            features.update({
                "heart_rate_mean": np.mean(heart_rates),
                "heart_rate_std": np.std(heart_rates),
                "heart_rate_min": np.min(heart_rates),
                "heart_rate_max": np.max(heart_rates),
            })

        # 2. Signal Characteristics
        features.update({
            "signal_mean": np.mean(ecg_cleaned),
            "signal_std": np.std(ecg_cleaned),
            "signal_skewness": skew(ecg_cleaned),
            "signal_kurtosis": kurtosis(ecg_cleaned),
            "signal_peak_to_peak": np.max(ecg_cleaned) - np.min(ecg_cleaned),
        })

        # 3. Frequency Domain Features
        f, psd = signal.welch(ecg_cleaned, fs)
        features.update({
            "total_power": np.sum(psd),
            "peak_frequency": f[np.argmax(psd)],
            "median_frequency": f[np.cumsum(psd) >= np.sum(psd)/2][0],
        })

        # 4. R-Peak Related Features
        if len(r_peaks) > 0:
            features.update({
                "r_peak_count": len(r_peaks),
                "r_peak_amplitude_mean": np.mean(ecg_cleaned[r_peaks]),
                "r_peak_amplitude_std": np.std(ecg_cleaned[r_peaks]),
            })

        # 5. Wave Characteristics - Robust Implementation
        try:
            # Delineate waves with fallback
            try:
                _, waves_info = nk.ecg_delineate(ecg_cleaned, info, sampling_rate=fs)

                # Manual extraction of wave characteristics
                wave_features = {}

                # QRS Complex Duration
                if 'ECG_R_Peaks' in waves_info and len(waves_info['ECG_R_Peaks']) > 0:
                    qrs_onsets = waves_info.get('ECG_Q_Onset', [])
                    qrs_offsets = waves_info.get('ECG_S_Offset', [])

                    if len(qrs_onsets) > 0 and len(qrs_offsets) > 0:
                        qrs_durations = [offset - onset for onset, offset in zip(qrs_onsets, qrs_offsets)]
                        wave_features['qrs_duration_mean'] = np.mean(qrs_durations)

                # P and T Wave Approximation
                p_onsets = waves_info.get('ECG_P_Onset', [])
                t_onsets = waves_info.get('ECG_T_Onset', [])

                if len(p_onsets) > 0:
                    wave_features['p_wave_onset_mean'] = np.mean(p_onsets)

                if len(t_onsets) > 0:
                    wave_features['t_wave_onset_mean'] = np.mean(t_onsets)

                features.update(wave_features)

            except Exception as wave_error:
                print(f"Wave feature extraction detailed error: {wave_error}")

        except Exception as wave_extraction_error:
            print(f"Comprehensive wave extraction error: {wave_extraction_error}")

        # 6. Heart Rate Variability - Robust Implementation
        try:
            # Convert R-peak indices to RR intervals
            rr_intervals = np.diff(r_peaks) / fs

            # Manual HRV calculations
            features.update({
                "hrv_sdnn": np.std(rr_intervals),
                "hrv_rmssd": np.sqrt(np.mean(np.square(np.diff(rr_intervals)))),
            })

            # Additional HRV metrics
            nn_diff = np.diff(rr_intervals)
            features.update({
                "hrv_pnn50": np.sum(np.abs(nn_diff) > 0.05) / len(nn_diff) * 100,
                "hrv_pnn20": np.sum(np.abs(nn_diff) > 0.02) / len(nn_diff) * 100,
            })

        except Exception as hrv_error:
            print(f"HRV feature extraction error: {hrv_error}")

        # 7. Signal Quality Indicators
        features.update({
            "baseline_wander": np.std(ecg_cleaned - filtered_signal),
            "signal_noise_ratio": np.max(ecg_cleaned) / np.std(ecg_cleaned),
        })

        return features

    except Exception as e:
        print(f"Comprehensive feature extraction critical error: {e}")
        raise

def process_ecg_data(vals_str, scale_factor=1000, fs=500, lowcut=0.5, highcut=50):
    """
    Robust ECG data processing pipeline with comprehensive feature extraction

    Parameters:
    - vals_str: Input ECG signal (string or list of values)
    - scale_factor: Scaling factor for signal normalization
    - fs: Sampling frequency
    - lowcut: Low cutoff frequency for bandpass filter
    - highcut: High cutoff frequency for bandpass filter

    Returns:
    - Dictionary of extracted ECG features
    """
    try:
        # Flexible input parsing
        if isinstance(vals_str, str):
            raw_signal = [float(i.strip()) for i in vals_str.strip('[]').split(',')]
        elif isinstance(vals_str, list):
            raw_signal = vals_str
        else:
            raise ValueError("Invalid input type. Expected string or list of values")

        # Ensure minimum signal length
        if len(raw_signal) < 250:  # At least 1 second of data
            raise ValueError(f"Signal too short. Minimum 250 points required. Current length: {len(raw_signal)}")

        # Limit signal length
        raw_signal = raw_signal[:fs*10]  # Limit to 10 seconds of data

        # Scale the signal
        scaled_signal = [x / scale_factor for x in raw_signal]

        # Apply bandpass filter
        filtered_signal = bandpass_filter(scaled_signal, lowcut, highcut, fs)

        # Extract comprehensive features
        features = extract_comprehensive_ecg_features(filtered_signal, fs=fs)

        return features

    except Exception as e:
        print(f"Error in ECG data processing: {e}")
        raise

def debug_ecg_processing(vals_str):
    """
    Comprehensive debugging for ECG data processing

    Parameters:
    - vals_str: Input ECG signal
    """
    # Convert to list and print details
    raw_signal = [float(i.strip()) for i in vals_str.strip('[]').split(',')]
    print("Signal details:")
    print("  Length:", len(raw_signal))
    print("  Min value:", min(raw_signal))
    print("  Max value:", max(raw_signal))
    print("  Mean value:", np.mean(raw_signal))

    # Process with different parameters
    try:
        features = process_ecg_data(raw_signal, scale_factor=1000, fs=250)
        print("\nExtracted Features:")
        for key, value in features.items():
            print(f"  {key}: {value}")
    except Exception as e:
        print(f"\nProcessing failed: {e}")
```

```python
import matplotlib.pyplot as plt

def plot_ecg_signal(vals_str, scale_factor=1000):
    """
    Plot the ECG signal to ensure it's correct.

    Parameters:
    - vals_str: String of raw ECG values (e.g., "[16000, 24000, ...]").
    - scale_factor: Factor to scale raw values into meaningful units. Default is 1000 (e.g., 1000/mV).

    Returns:
    - None (displays a plot).
    """
    # Convert the string to a list of integers
    raw_signal = [int(i) for i in vals_str.strip('[]').split(',')]

    # Scale the signal
    scaled_signal = [x / scale_factor for x in raw_signal]

    # Plot the signal
    plt.figure(figsize=(10, 4))
    plt.plot(scaled_signal, label="ECG Signal", color='blue')
    plt.title("ECG Signal")
    plt.xlabel("Samples")
    plt.ylabel("Amplitude (mV)")
    plt.grid(True)
    plt.legend()
    plt.show()

plot_ecg_signal(vals)
```
